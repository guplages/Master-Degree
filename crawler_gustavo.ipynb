```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url1 = "https://www.cia.gov/library/readingroom/collection/intelligence-policy-and-politics-dci-white-house-and-congress?page="

links = []

for page_id in range(68):
    response = requests.get(url1+str(page_id))
    soup = BeautifulSoup(response.text, "html.parser")
    for link in soup.findAll("a"):
        try:
            if 'pdf' in link['href']:
                links.append(link['href'])
                print(link['href'])
        except:
            pass
```

```python
for num, link in enumerate(links):
    name = 'file{}_'.format(num) + link.split('/docs/')[1]
    print(name)
    urllib.request.urlretrieve(link, name)
    time.sleep(1)
```

```python
url2 = "https://www.cia.gov/library/readingroom/search/site?page={}&f%5B0%5D=im_field_taxonomy_nic_function%3A2"

links = []

for page_id in range(6):
    response = requests.get(url2.format(page_id))
    soup = BeautifulSoup(response.text, "html.parser")
    for link in soup.findAll("a"):
        try:
            if '/readingroom/document/' in link['href']:
                links.append(link['href'])
                print(link['href'])
        except:
            pass
```

```python
for num, link in enumerate(links):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, "html.parser")
    linkspage = soup.findAll("a")
    for linkpage in linkspage:
        try:
            if 'pdf' in linkpage['href']:
                name = 'file_2_{}_'.format(num) + linkpage['href'].split('/docs/')[1]
                print(name)
                urllib.request.urlretrieve(linkpage['href'], name)
                time.sleep(1)
        except:
            pass
```
